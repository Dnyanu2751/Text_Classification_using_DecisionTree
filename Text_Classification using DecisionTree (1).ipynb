```python
import pandas as pd
A = pd.read_csv("C:/users/HP/Downloads/Restaurant_Reviews.tsv",sep="\t")
```

# Missing data treatment

```python
A.isna().sum()
```

# Preprocessing

```python
from re import sub
```

```python
q = []
for i in A.Review:
    w = sub("[^a-zA-Z0-9 ]","",i)
    w = w.lower()
    q.append(w)
```

```python
q[0:5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(q).toarray()
```

```python
cv.get_feature_names()
```

```python
Xnew = pd.DataFrame(X,columns=cv.get_feature_names())
```

```python
Xnew[["this"]]
```

```python
Y = A.Liked
```

# Training and testing

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=21)
```

# Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=21,max_depth=3)
model = dtc.fit(xtrain,ytrain)
pred = model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest,pred)
```

# Review classification using user input

```python
rev = input("Enter your review: ")
inp_text = sub("[^A-Z0-9 ]","",rev.lower())
X1 = cv.transform([inp_text]).toarray()
e = model.predict(X1)[0]
if(e == 0):
    print("Disliked")
else:
    print("Liked")
```
